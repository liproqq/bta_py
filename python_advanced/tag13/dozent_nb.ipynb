# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Klasse für das Neuron 

```python
class LinearRegression:
    def __init__(self, eta=0.0001, epochs=25):
        self.eta = eta
        self.epochs = epochs

    def fit(self, X, y):

        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        self.cost = []

        for i in range(self.epochs):
            prediction = self.predict(X).reshape(y.shape)

            errors = y - prediction

            mse = (errors**2).mean()

            self.cost.append(mse)

            errors = errors.ravel()
            self.weights += self.eta * X.T.dot(errors)
            self.bias += self.eta * errors.sum()

        return self

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

#### Hauptprogramm

```python
%matplotlib inline

X = 2 * np.random.rand(1000, 1)
y = 5 + 2 * X + np.random.randn(1000, 1)

plt.scatter(X, y)
for i in range(100):
    lr = LinearRegression(epochs=i, eta=0.0001)
    lr.fit(X, y)
    plt.plot(X, lr.predict(X), label=f'Epoch: {i}')

# plt.legend()
plt.show()

plt.plot(range(1, lr.epochs + 1), lr.cost)
plt.ylabel('MSE')
plt.xlabel('Epoch')
plt.show()

print(f'y = {lr.weights[0]:.2f}*x + {lr.bias:.2f}')
```
