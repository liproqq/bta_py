```python
import cv2
import numpy as np
import pickle
import matplotlib.pyplot as plt
from pathlib import Path
import os
from sklearn.utils import shuffle
```

```python
folder = Path().resolve() / "PetImages"
```

```python
for category in ["Cat", "Dog"]:
    path = folder / category
    for img_path in path.iterdir():
        img_array = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)

        plt.imshow(img_array, cmap="gray")
        plt.show()

        img_array = cv2.resize(img_array, (100, 100))
        plt.imshow(img_array, cmap="gray")
        plt.show()
        break
```

```python
def create_dataset(DATADIR="PetImages", CATEGORIES=["Cat", "Dog"], IMG_SIZE=100):
    features = []
    target = []
    errors = []

    for label, category in enumerate(CATEGORIES):
        folder = Path().resolve() / DATADIR
        path = folder / category
        for img_path in path.iterdir():
            try:
                img_array = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)
                img_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))

                features.append(img_array)
                target.append(label)
            except Exception as e:
                errors.append(img_path)
                print(e)
    return shuffle(np.array(features), np.array(target))
```

```python
feat, targ = create_dataset()
```
